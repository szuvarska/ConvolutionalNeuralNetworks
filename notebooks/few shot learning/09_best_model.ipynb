```python
from src.few_shot_learning import load_cinic10, calculate_accuracy, plot_confusion_matrix, FewShotConvNeXt, train_few_shot_convnext, set_seed
from src.plots_functions import multiple_runs_with_uncertainty_band
```

```python
set_seed(213)
```

```python
data_dir = "../../data"
```

```python
train_dataloader = load_cinic10(data_dir, few_shot_per_class=100)
test_dataloader = load_cinic10(data_dir, few_shot_per_class=100, split="test")
```

```python
metrics_list = []
for _ in range(5):
    model = FewShotConvNeXt()
    metrics, time = train_few_shot_convnext(model, train_dataloader, test_dataloader, epochs=20, lr=0.0001, optimizer='adam', scheduling=True)
    metrics_list.append(metrics)
```

```python
experiment_name = "FewShotConvNeXt with lr=0.0001, Adam optimizer and scheduling and no dropout"
multiple_runs_with_uncertainty_band(metrics_list, f"Accuracy for {experiment_name}", f"Loss for {experiment_name}")
```

```python
plot_confusion_matrix(model, test_dataloader, title=f"Confusion matrix for {experiment_name}")
```
